```python
import tkinter as tk
import numpy as np
import pandas as pd
from string import punctuation
import tkinter.messagebox as messagebox

window = tk.Tk()
window.title('David - Your Education Chatbot')
window.geometry('1000x800')

var1 = tk.StringVar()##dialog 1
stg1 = tk.StringVar()##dialod 2
var2 = tk.StringVar()##dialog 1
stg2 = tk.StringVar()##dialod 2
var3 = tk.StringVar()##dialog 1
stg3 = tk.StringVar()##dialod 2
var4 = tk.StringVar()##dialog 1
stg4 = tk.StringVar()##dialod 2
var5 = tk.StringVar()##dialog 1
stg5 = tk.StringVar()##dialod 2

D_name = format('David:',)
U_name = format('User:',)

v = []#list1 for dialog1, store the dialog1
s = []#list2 for dialog2, store the dialog2

## Sample Model
def sample_model(e_stg):
    if e_stg == 'Where is the apple?':
        prob = 0.9
        answer = 'Bedroom'
    elif e_stg == 'Where is Lily?':
        prob = 0.75
        answer = 'Kitchen'
    elif e_stg == 'What color is Greg?':
        prob = 0.55
        answer = 'White'
    elif e_stg == 'Is Sam in the kitchen?':
        prob = 0.96
        answer = 'Yes'
    elif e_stg == 'Where is the milk?':
        prob = 0.35
        answer = 'Livinroom'
    else:
        prob = 0.05
        answer = 'Something irrelevant'
    return prob, answer


g_a = ['hello','hi','howdy']
ga_r = 'Hello!'
g_b = 'your name'
gb_r = 'My name is David.'
g_c = 'nice to meet you'
gc_r = 'Nice to meet you, too!'
g_d = 'can you do'
gd_r = 'Well, I am your best friend in readings, you can ask me some question about the story.'
g_e = 'get start'
ge_r = 'Sure.'
g_f = 'great job'
gf_r = 'Thank you! I will continue to do my best.'
g_g = 'thank'
gg_r = 'You\'re welcome!'
g_end = 'the nlp'
gend_r = 'Oh, the Natural Language Processing course? That was fantastic! I mean, without Professor Gasan Elkhodari\'s teaching and your fellow\'s hard work, I don\'t even exist. Thank you all!'

e_stg=''
i = 0   

###functions
def hitButton():
    global v,s,vb,sb,e_stg,i
    e_stg = e.get()
    if e_stg == '':
        messagebox.showerror(title='ERROR',message='There is nothing, try to ask something to me!')
    else:
        s.append(e_stg)
        stg5.set('User: '+s[-1])
        if e_stg.strip(punctuation).lower() in g_a:
    #         var.set(ga_r)
            v.append(ga_r)
        elif g_b in e_stg.strip(punctuation).lower():
    #         var.set(gb_r)
            v.append(gb_r)
        elif g_c in e_stg.strip(punctuation).lower():
    #         var.set(gc_r)
            v.append(gc_r)
        elif g_d in e_stg.strip(punctuation).lower():
    #         var.set(gd_r)
            v.append(gd_r)
        elif g_e in e_stg.strip(punctuation).lower():
    #         var.set(ge_r)
            v.append(ge_r)
        elif g_f in e_stg.strip(punctuation).lower():
    #         var.set(gf_r)
            v.append(gf_r)
        elif g_g in e_stg.strip(punctuation).lower():
    #         var.set(gg_r)
            v.append(gg_r)
        elif g_end in e_stg.strip(punctuation).lower():
    #         var.set(gend_r)
            v.append(gend_r)
        else:
            prob,answer = sample_model(e_stg)
            if prob >= 0.9:
                strg = 'I am pretty sure the answer is '+answer.lower()+'.'
                v.append(strg)
            elif prob >=0.7 and prob <0.9:
                strg = 'I think the answer is '+answer.lower()+' probably.'
                v.append(strg)
            elif prob >=0.5 and prob <0.7:
                strg = 'I am not sure about this, but the answer is '+answer.lower()+' possibly.'
                v.append(strg)
            elif prob >=0.3 and prob <0.5:
                strg = 'I really don\'t know, but I guess the answer would be '+answer+'.'
                v.append(strg)
            else:
                strg = 'I am sorry, I have no answer for this right now, please try something else to ask.'
                v.append(strg)

        var5.set('David: '+v[-1])

        i += 1
        if i == 1:
            var3.set('')
            stg3.set('')
            var4.set('')
            stg4.set('')


        elif i == 2:
            var3.set('')
            stg3.set('')
            var4.set('David: '+v[0])
            stg4.set('User: '+s[0])

        elif i == 3:
            var3.set('David: '+v[0])
            stg3.set('User: '+s[0])
            var4.set('David: '+v[1])
            stg4.set('User: '+s[1])

        elif i == 4:
            var2.set('David: '+v[0])
            stg2.set('User: '+s[0])
            var3.set('David: '+v[1])
            stg3.set('User: '+s[1])
            var4.set('David: '+v[2])
            stg4.set('User: '+s[2])

        elif i == 5:
            var1.set('David: '+v[0])
            stg1.set('User: '+s[0])
            var2.set('David: '+v[1])
            stg2.set('User: '+s[1])
            var3.set('David: '+v[2])
            stg3.set('User: '+s[2])
            var4.set('David: '+v[3])
            stg4.set('User: '+s[3])

        else:
            s.remove(s[0])
            v.remove(v[0])
            var1.set('David: '+v[0])
            stg1.set('User: '+s[0])
            var2.set('David: '+v[1])
            stg2.set('User: '+s[1])
            var3.set('David: '+v[2])
            stg3.set('User: '+s[2])
            var4.set('David: '+v[3])
            stg4.set('User: '+s[3])

        e.delete(0,500)
        
        
        
width_test = 75
font_test = 15
x_test = 75
wraplength_test = 600
l1_1 = tk.Label(window, textvariable = stg1, width = width_test, font=('Times new roman',font_test), justify = 'right', anchor = 'e',wraplength = wraplength_test)
l2_1 = tk.Label(window, textvariable = var1, width = width_test, font=('Times new roman',font_test), justify = 'left', anchor = 'w',wraplength = wraplength_test)
l1_2 = tk.Label(window, textvariable = stg2, width = width_test, font=('Times new roman',font_test), justify = 'right', anchor = 'e',wraplength = wraplength_test)
l2_2 = tk.Label(window, textvariable = var2, width = width_test, font=('Times new roman',font_test), justify = 'left', anchor = 'w',wraplength = wraplength_test)
l1_3 = tk.Label(window, textvariable = stg3, width = width_test, font=('Times new roman',font_test), justify = 'right', anchor = 'e',wraplength = wraplength_test)
l2_3 = tk.Label(window, textvariable = var3, width = width_test, font=('Times new roman',font_test), justify = 'left', anchor = 'w',wraplength = wraplength_test)
l1_4 = tk.Label(window, textvariable = stg4, width = width_test, font=('Times new roman',font_test), justify = 'right', anchor = 'e',wraplength = wraplength_test)
l2_4 = tk.Label(window, textvariable = var4, width = width_test, font=('Times new roman',font_test), justify = 'left', anchor = 'w',wraplength = wraplength_test)
l1_5 = tk.Label(window, textvariable = stg5, width = width_test, font=('Times new roman',font_test), justify = 'right', anchor = 'e',wraplength = wraplength_test)
l2_5 = tk.Label(window, textvariable = var5, width = width_test, font=('Times new roman',font_test), justify = 'left', anchor = 'w',wraplength = wraplength_test)

l1_1.place(x = x_test,y = 10)
l2_1.place(x = x_test, y = 80)
l1_2.place(x = x_test,y = 150)
l2_2.place(x = x_test, y = 220)
l1_3.place(x = x_test,y = 290)
l2_3.place(x = x_test, y = 360)
l1_4.place(x = x_test,y = 430)
l2_4.place(x = x_test, y = 500)
l1_5.place(x = x_test,y = 570)
l2_5.place(x = x_test, y = 640)

##### click the button, have the event OR you can just click the ENTER key
### button events function:
def onMouse(event):
    b.configure(bg='green', fg='white')
def outMouse(event):
    b.configure(bg='lightblue',fg='black')
def returnEvent(event):
    hitButton()
    



e = tk.Entry(window, show=None, font=('Times new roman',20),width=50,bd = 3)
e.place(x=90,y=730)
b = tk.Button(window, text='ask', bg='lightblue',fg='black', font=('Times new roman', 16)
               ,command=hitButton)
b.place(x=840,y=729)

b.bind('<Enter>',onMouse)
b.bind('<Leave>',outMouse)
e.bind('<Return>',returnEvent)


window.mainloop()
```
